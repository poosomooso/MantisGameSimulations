```python
from mantis_game import play, Player, GameLogger
import custom_players
import matplotlib.pyplot as plt
```

# Always Score

```python
num_players = 5
players = []
for i in range(num_players):
    players.append(custom_players.AlwaysScore())

always_score_logger = play(specified_players=players, num_players=num_players)
```

```python
always_score_logger.actions
```

```python
def plot_points(gl):
    plt.figure()
    plt.title("Num points")
    for p in range(num_players):
        points = gl.num_points[p]
        plt.plot(range(len(points)), points)
    plt.show()

def plot_cards(gl):
    plt.figure()
    plt.title("Num cards")
    for p in range(num_players):
        points = gl.num_cards[p]
        plt.plot(range(len(points)), points)
    plt.show()

    
```

```python
plot_points(always_score_logger)
plot_cards(always_score_logger)
```

```python
score_color = "#2191FB"
steal_color = "#BA274A"
def plot_actions_vs_points(gl):
    fig = plt.figure()
    plt.title("Num points vs action")
    for p in range(num_players):
        points = gl.num_points[p]
        actions = gl.actions[p]
        plt.plot(range(len(points)), points)
        score_pts = [ (turn, points) for turn, points in enumerate(points) if actions[turn]["action"] == "score"]
        score_pts = list(zip(*score_pts))
        steal_pts = [ (turn, points) for turn, points in enumerate(points) if actions[turn]["action"] == "steal"]
        steal_pts = list(zip(*steal_pts))
        score_artist = None
        steal_artist = None
        if len(score_pts) > 0:
            score_artist = plt.scatter(score_pts[0], score_pts[1], color=score_color, marker="P", label="score")
        if len(steal_pts) > 0:
            steal_artist = plt.scatter(steal_pts[0], steal_pts[1], color=steal_color, marker="v", label="steal")
    plt.legend([score_artist, steal_artist], ["scored", "stole"])
    
    plt.show()

def plot_actions_vs_cards(gl):
    fig = plt.figure()
    plt.title("Num cards vs action")
    for p in range(num_players):
        points = gl.num_cards[p]
        actions = gl.actions[p]
        plt.plot(range(len(points)), points)
        score_pts = [ (turn, points) for turn, points in enumerate(points) if actions[turn]["action"] == "score"]
        score_pts = list(zip(*score_pts))
        steal_pts = [ (turn, points) for turn, points in enumerate(points) if actions[turn]["action"] == "steal"]
        steal_pts = list(zip(*steal_pts))
        score_artist = None
        steal_artist = None
        if len(score_pts) > 0:
            score_artist = plt.scatter(score_pts[0], score_pts[1], color=score_color, marker="P", label="score")
        if len(steal_pts) > 0:
            steal_artist = plt.scatter(steal_pts[0], steal_pts[1], color=steal_color, marker="v", label="steal")
    plt.legend([score_artist, steal_artist], ["scored", "stole"])
    
    plt.show()
```

```python
plot_actions_vs_points(always_score_logger)
plot_actions_vs_cards(always_score_logger)
```

# Always Steal Random
Always steal from someone who viably has a matching card

```python
num_players = 5
players = []
for i in range(num_players):
    players.append(custom_players.AlwaysStealRandom())

always_steal_logger = play(specified_players=players, num_players=num_players)
```

```python
plot_actions_vs_points(always_steal_logger)
plot_actions_vs_cards(always_steal_logger)
```

# Always Steal Max Matching ratio

Always steal from someone who has the highest percentage of cards that might match

```python
num_players = 5
players = []
for i in range(num_players):
    players.append(custom_players.AlwaysStealMaxMatchingRatio())

always_steal_logger = play(specified_players=players, num_players=num_players)

plot_actions_vs_points(always_steal_logger)
plot_actions_vs_cards(always_steal_logger)
```

# Always Steal Max Diversity Ratio

Always steal from someone who has the highest probability of having the color of the card

```python
num_players = 5
players = []
for i in range(num_players):
    players.append(custom_players.AlwaysStealMaxDiversity())

always_steal_logger = play(specified_players=players, num_players=num_players)

plot_actions_vs_points(always_steal_logger)
plot_actions_vs_cards(always_steal_logger)
```

# Safety Steal
